```python

# Definir una función para aplicar a cada fila del DataFrame
def evaluar_fila(row):
    if (row['Enero'] == 0) or (row['Febrero'] == 0) or (row['Marzo'] == 0) or (row['Abril']) == 0:
        return 'Registros Incompletos'
    else:
        return 'Good'
def evaluar_fila2(row):
    if (row['Enero']== 0) and (row['Febrero']==0) and (row['Marzo']==0) and (row['Abril']) == 0:
        return 'Sin Registros'
    else:
        return 'Registros Incompletos'
# Aplicar la función a cada fila del DataFrame
centros_respel=dfpeligrosos.copy()
centros_respel['Resultado'] = centros_respel.apply(evaluar_fila, axis=1)
centros_respel=centros_respel.loc[(centros_respel['Resultado']== 'Registros Incompletos' ) ]
centros_respel['Resultado']=centros_respel.apply(evaluar_fila2,axis=1)
for i in range(centros_respel['Resultado'].count()):
     centros_respel['Indicador']=centros_respel['Resultado'].apply(lambda x: 'Residuos')
```
